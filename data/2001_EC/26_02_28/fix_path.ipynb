```python
import sys, os, importlib.util

print("=== 1. Python executable ===")
print(sys.executable)

print("\n=== 2. sys.path ===")
print('\n'.join(sys.path))

print("\n=== 3. PYTHONPATH ===")
print(os.environ.get('PYTHONPATH', 'NOT SET'))

print("\n=== 4. utils folder contents ===")
utils_path = '/Users/robin/Documents/Astrophysical observation/OUKA2026/utils'
try:
    print(os.listdir(utils_path))
except Exception as e:
    print(f"ERROR: {e}")

print("\n=== 5. Direct import attempt ===")
try:
    spec = importlib.util.spec_from_file_location("utils", utils_path + "/__init__.py")
    mod = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(mod)
    print(f"Success: {mod}")
except Exception as e:
    print(f"ERROR: {e}")
```
