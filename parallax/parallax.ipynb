```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# TN
long1 = -7.8664
lat1 = 31.2061

# TS
long2 = -70.7394
lat2 = -29.2546

R = 6372 # km

# Convert to 3D cartesian coordinates
def to_xyz(lat, lon):
    
    lat = lat*np.pi/180
    lon = lon*np.pi/180
    
    x = R * np.cos(lat) * np.cos(lon)
    y = R * np.cos(lat) * np.sin(lon)
    z = R * np.sin(lat)
    
    return np.array([x, y, z])

coord1 = to_xyz(lat1, long1)
coord2 = to_xyz(lat2, long2)

d = np.sqrt((coord1[0] - coord2[0])**2 + (coord1[1] - coord2[1])**2 + (coord1[2] - coord2[2])**2)
print(d)
```
